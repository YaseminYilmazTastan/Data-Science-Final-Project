```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading data from csv file
diamond = pd.read_csv("Diamonds Prices2022.csv")
diamond.head()
```

```python
diamond.info()
```

```python
diamond.describe()
```

```python
#drop first column from data set
diamond= diamond.drop(["Unnamed: 0"], axis=1)
```

```python
diamond.head(10)
```

```python
#identify some outlines in our data 
sns.pairplot(diamond)
```

```python
#with group by method ,we can see  how price change  according  to cut and color
grup = diamond.groupby(['cut' , 'color'])
```

```python
grup["price"].mean().plot(kind ="bar")
```

```python
#we can see how price and carat relationship
diamond.plot(kind="scatter",x= "price", y= "carat")
```

```python
#we want to see our data distribution
diamond.columns
```

```python
diamond["carat"].value_counts()
```

```python
sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",data=diamond)
```
